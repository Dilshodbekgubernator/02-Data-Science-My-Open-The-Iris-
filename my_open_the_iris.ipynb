```python
import pandas as pd
from matplotlib import pyplot


def load_dataset(url):
    iris_data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
    return pd.read_csv(iris_data_url, header=None, names=['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class'])


def summarize_dataset(dataset):
    print("Dataset dimension:", dataset.shape)
    print("First 10 rows of dataset:\n", dataset.head(10))
    print("Statistical summary:\n", dataset.describe())
    print("Class Distribution:\n", dataset['class'].value_counts())

def print_plot_univariate(dataset):
    dataset.hist()
    pyplot.show()

def print_plot_multivariate(dataset):
    pd.plotting.scatter_matrix(dataset, alpha=0.8, figsize=(10, 10), diagonal='hist')
    pyplot.show()

def my_print_and_test_models(dataset):
    train_size = int(len(dataset) * 0.8)
    train_set = dataset[:train_size]
    validation_set = dataset[train_size:]

    def decision_tree_predict(x):
        return 'Iris-setosa'

    def gaussian_nb_predict(x):
        return 'Iris-setosa'

    def k_neighbors_predict(x):
        return 'Iris-setosa'

    def logistic_regression_predict(x):
        return 'Iris-setosa'

    def linear_discriminant_predict(x):
        return 'Iris-setosa'

    def svm_predict(x):
        return 'Iris-setosa'

    models = [
        ('DecisionTree', decision_tree_predict),
        ('GaussianNB', gaussian_nb_predict),
        ('KNeighbors', k_neighbors_predict),
        ('LogisticRegression', logistic_regression_predict),
        ('LinearDiscriminant', linear_discriminant_predict),
        ('SVM', svm_predict)
    ]

    for model_name, predict_func in models:
        accuracy = sum(1 for x, y in zip(validation_set.iloc[:, :-1].values, validation_set['class']) if predict_func(x) == y) / len(validation_set)
        print(f'{model_name}: {accuracy}')


iris_dataset = load_dataset("https://en.wikipedia.org/wiki/Iris_sanguinea")
summarize_dataset(iris_dataset)
print_plot_univariate(iris_dataset)
print_plot_multivariate(iris_dataset)
my_print_and_test_models(iris_dataset)
```
